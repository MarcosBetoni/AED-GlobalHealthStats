# Importando as bibliotecas

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

# Carregando o conjunto de dados (exemplo: Titanic dataset)


```python
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)
```

```python
print("\n Informações sobre o dataframe:")
print(df.info())
```

# Estatísticas descritivas


```python
print("\nEstatísticas descritivas:")
display(df.describe())
```

# Visualizando a distribuição da idade dos passageiros


```python
df['FareLog'] = np.log2(df['Fare'])
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x="Fare", bins=100, kde=True)
plt.title("Distribuição da Fare dos Passageiros")
plt.xlabel("Fare")
plt.ylabel("Frequência")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df[df['Survived']==1], x="Age", bins=30, kde=True)
plt.title("Distribuição da Idade dos Passageiros Sobreviventes")
plt.xlabel("Idade")
plt.ylabel("Frequência")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df[df['Survived']==0], x="Age", bins=30, kde=True)
plt.title("Distribuição da Idade dos Passageiros Não Sobreviventes")
plt.xlabel("Idade")
plt.ylabel("Frequência")
plt.show()
```

# Visualizando a contagem de sobreviventes por sexo


```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x="Survived", hue="Sex")
plt.title("Contagem de Sobreviventes por Sexo")
plt.xlabel("Sobreviveu?")
plt.ylabel("Contagem")
plt.legend(title="Sexo")
plt.show()
```

```python

df_sexo = df.groupby(['Survived']).count() / df.shape[0]
df_sexo['PassengerId']
```

# Visualizando a distribuição da tarifa dos passageiros


```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x="Fare", bins=60, kde=True)
plt.title("Distribuição da Tarifa dos Passageiros")
plt.xlabel("Tarifa")
plt.ylabel("Frequência")
plt.show()
```

# Visualizando a contagem de sobreviventes por classe


```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x="Survived", hue="Pclass")
plt.title("Contagem de Sobreviventes por Classe")
plt.xlabel("Sobreviveu?")
plt.ylabel("Contagem")
plt.legend(title="Classe")
plt.show()
```

# Visualizando a relação entre idade, tarifa e sobrevivência

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x="Age", y="Fare", hue="Survived", palette="Set1")
plt.title("Relação entre Idade, Tarifa e Sobrevivência")
plt.xlabel("Idade")
plt.ylabel("Tarifa")
plt.legend(title="Sobreviveu?")
plt.show()
```

# Verificando a correlação entre as variáveis numéricas


```python
plt.figure(figsize=(10, 6))
sns.heatmap(df[['Survived','Age','Pclass','Fare']].corr(), annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlação entre Variáveis Numéricas")
plt.show()
```
